```python
import matplotlib.pyplot as plt
import seaborn as sns

# Reemplazar valores infinitos
# Si se encuentran valores infinitos, reemplácelos con NaN,
# y luego manejarlos como manejaría los valores faltantes.
data.replace([np.inf, -np.inf], np.nan, inplace=True)
data.dropna(inplace=True)

# Establecer estilo de trama
sns.set(style="whitegrid")

# Histogramas de edad, horas de estudio, puntuaciones de Python y DB
fig, ax = plt.subplots(2, 2, figsize=(12, 8))

sns.histplot(data=data['price'], kde=True, ax=ax[0, 0]).set(title='price Distribution')
sns.histplot(data=data['rating'], kde=True, ax=ax[0, 1]).set(title='rating Distribution')
sns.histplot(data=data['units_sold'], kde=True, ax=ax[1, 0]).set(title='units_sold Distribution')
sns.histplot(data=data['inventory_total'], kde=True, ax=ax[1, 1]).set(title='inventory_total Distribution')

plt.tight_layout()
plt.show()
```

```python
# Matriz de correlación
corr_matrix = data [['price', 'rating', 'units_sold', 'inventory_total']].corr()

# Trazar el mapa de calor de la matriz de correlación
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, linewidths=0.2)
plt.title("Correlation Matrix")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot to show relationship
plt.figure(figsize=(10,6))
sns.scatterplot(x=data2['totalunitssold'], y=data2['rating'])
plt.title("Relationship between totalunitssold and rating")
plt.xlabel("totalunitssold")
plt.ylabel("rating")
plt.show()

# Correlation coefficient
correlation = data2['totalunitssold'].corr(data2['rating'])
print()
print(' * '*20)
print(f'The Correlationship between totalunitssold and rating is: {correlation:.2f}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.boxplot(x=data2['totalunitssold'], y=data2['meanproductprices'])
plt.title("Relacion de total de unidades de vendidas con el promedio de precio de productos")
plt.xlabel("totalunitssold")
plt.ylabel("meanproductprices")
plt.xticks(rotation=45)
plt.show()


prom_unidadesvendidas = data2.groupby('totalunitssold')['meanproductprices'].mean()
print('Average (Mean) promedio de precio por unidades vendidas')
print(prom_unidadesvendidas)

print(' * '*20)

education_python_median = data2.groupby('totalunitssold')['meanproductprices'].median()
print('Promedio de unidades vendidas por precio')
print(prom_unidadesvendidas)
```

Los precios varían entre 4.81 y 13.50, con la mayoría de los valores alrededor de los 7-9.
El precio promedio parece estar relativamente disperso, lo que sugiere que hay productos con precios tanto bajos como altos, dependiendo del volumen de unidades vendidas.
La relación entre el precio y el volumen de ventas no parece ser estrictamente lineal, lo que indica que no necesariamente un precio más alto está asociado con menos ventas o viceversa.


Si intentamos comparar precio promedio por unidades vendidas y unidades vendidas por precio, podríamos estar observando patrones:

Precio más bajo, mayor volumen de ventas: En algunos casos, productos con precios más bajos podrían estar asociados con mayores volúmenes de ventas.

Precio más alto, menos ventas: Los productos con precios más altos pueden tener un volumen de ventas menor, pero esto no siempre es cierto, ya que en algunos casos, un precio más alto puede estar asociado a productos de mayor calidad o a nichos de mercado específicos, lo que podría mantener las ventas estables.

Estrategias de precios: Los datos podrían sugerir que algunos productos de precio intermedio  parecen estar vendiendo bien, mientras que los precios demasiado bajos o demasiado altos podrían no ser tan efectivos.
