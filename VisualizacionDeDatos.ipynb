```python

plt.figure(figsize=(10, 6))
plt.plot(stock_prices['Date'], stock_prices['JPM'], label='JP Morgan')
plt.plot(stock_prices['Date'], stock_prices['PG'], label='P&G')
plt.plot(stock_prices['Date'], stock_prices['AAPL'], label='Apple')
plt.plot(stock_prices['Date'], stock_prices['UAL'], label='United Airlines')


plt.xlabel('Fecha')
plt.ylabel('Precio de la acci√≥n')
plt.title('Precios diarios de acciones')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python

fig, axes = plt.subplots(2, 2, figsize=(12, 8))


axes[0, 0].plot(stock_prices['Date'], stock_prices['JPM'], color='blue')
axes[0, 0].set_title('JP Morgan')
axes[0, 1].plot(stock_prices['Date'], stock_prices['PG'], color='green')
axes[0, 1].set_title('Procter & Gamble')
axes[1, 0].plot(stock_prices['Date'], stock_prices['AAPL'], color='orange')
axes[1, 0].set_title('Apple')
axes[1, 1].plot(stock_prices['Date'], stock_prices['UAL'], color='red')
axes[1, 1].set_title('United Airlines')


for ax in axes.flat:
    ax.set(xlabel='Fecha', ylabel='Precio de la acci√≥n')
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python

fig, axes = plt.subplots(1, 4, figsize=(16, 4))


axes[0].plot(stock_prices['Date'], stock_prices['JPM'], color='blue')
axes[0].set_title('JP Morgan')
axes[1].plot(stock_prices['Date'], stock_prices['PG'], color='green')
axes[1].set_title('P&G')
axes[2].plot(stock_prices['Date'], stock_prices['AAPL'], color='orange')
axes[2].set_title('Apple')
axes[3].plot(stock_prices['Date'], stock_prices['UAL'], color='red')
axes[3].set_title('United Airlines')


for ax in axes:
    ax.set(xlabel='Fecha', ylabel='Precio de la acci√≥n')
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(8, 6))
plt.scatter(stock_returns['AAPL'], stock_returns['JPM'], alpha=0.5)
plt.xlabel('Retornos de Apple')
plt.ylabel('Retornos de JP Morgan')
plt.title('Diagrama de dispersi√≥n entre los retornos de Apple y JP Morgan')
plt.show()
```

```python
import seaborn as sns


plt.figure(figsize=(8, 6))
sns.scatterplot(data=stock_returns, x='AAPL', y='JPM')
plt.title('Diagrama de dispersi√≥n entre los retornos de Apple y JP Morgan (Seaborn)')
plt.show()
```

```python

labels = ['AAPL', 'JP Morgan', 'P&G', 'United Airlines']
sizes = [70, 10, 10, 10]
explode = (0.1, 0, 0, 0)


plt.figure(figsize=(7, 7))
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Asignaciones de activos')
plt.show()
```
