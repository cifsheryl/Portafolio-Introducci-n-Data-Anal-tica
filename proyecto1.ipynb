**DATA WRANGILING**

```python
# Crear una nueva variable que sea la relación entre salario y valor de mercado
players_16['salario_valor_ratio'] = players_16['wage_eur'] / players_16['value_eur']

# Mostrar las primeras filas para verificar
players_16[['short_name', 'wage_eur', 'value_eur', 'salario_valor_ratio']].head()
```

```python
# Eliminar las filas con NaN o inf en la columna 'salario_valor_ratio'
players_16_clean = players_16.dropna(subset=['salario_valor_ratio'])

# Verificar si se han eliminado correctamente las filas
print(players_16_clean['salario_valor_ratio'].isna().sum())

```

```python
print(players_16['salario_valor_ratio'].describe())
```

La media de la relación salario/valor es 0.0100, lo que sugiere que, en promedio, el salario de los jugadores es aproximadamente el 1% de su valor de mercado.
La desviación estándar es 0.0067, podemos considerar una variabilidad moderada en la relación salario/valor entre los jugadores.
El valor mínimo es 0.0002, lo que sugiere que hay jugadores cuyo salario es una fracción muy pequeña de su valor de mercado.
El valor máximo es 0.15, lo que indica que hay jugadores cuyo salario es significativamente mayor en comparación con su valor de mercado.
La relación salario/valor se distribuye de manera que el 25% de los jugadores tienen una relación menor a 0.0057, el 50% a 0.0091, y el 75% a 0.0128.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# histograma de la relación salario/valor
plt.figure(figsize=(10, 6))
sns.histplot(players_16_clean['salario_valor_ratio'], bins=50, kde=True, color='blue')


plt.title('Distribución de la Relación Salario/Valor', fontsize=16)
plt.xlabel('Relación Salario/Valor', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)


plt.show()
```

```python
# Jugadores con la mejor relación salario/valor
top_salario_valor = players_16_clean.sort_values('salario_valor_ratio', ascending=False).head(10)

# Jugadores con la peor relación salario/valor
bottom_salario_valor = players_16_clean.sort_values('salario_valor_ratio').head(10)


print("Mejor relación salario/valor:")
print(top_salario_valor[['short_name', 'salario_valor_ratio', 'wage_eur', 'value_eur']])

print("\nPeor relación salario/valor:")
print(bottom_salario_valor[['short_name', 'salario_valor_ratio', 'wage_eur', 'value_eur']])
```

```python
# correlación entre salario y valor de mercado
correlation = players_16_clean[['wage_eur', 'value_eur']].corr()

print("Correlación entre Salario y Valor de Mercado:")
print(correlation)
```

```python
# Agrupar por país y calcular la media de la relación salario/valor
salario_valor_por_pais = players_16_clean.groupby('nationality')['salario_valor_ratio'].mean().sort_values(ascending=False)

# Mostrar los 10 países con mayor relación salario/valor
print(salario_valor_por_pais.head(10))

# Crear un gráfico de barras para la relación salario/valor por país
plt.figure(figsize=(12, 8))
salario_valor_por_pais.head(10).plot(kind='bar', color='purple')

# Título y etiquetas
plt.title('Top 10 Países con Mayor Relación Salario/Valor', fontsize=16)
plt.xlabel('País', fontsize=12)
plt.ylabel('Relación Salario/Valor', fontsize=12)

# Mostrar el gráfico
plt.show()
```

```python
# Agrupar por posición y calcular la media de la relación salario/valor
salario_valor_por_posicion = players_16_clean.groupby('player_positions')['salario_valor_ratio'].mean().sort_values(ascending=False)

# Mostrar las 10 posiciones con mayor relación salario/valor
print(salario_valor_por_posicion.head(10))

# Crear un gráfico de barras para la relación salario/valor por posición
plt.figure(figsize=(12, 8))
salario_valor_por_posicion.head(10).plot(kind='bar', color='orange')

# Título y etiquetas
plt.title('Top 10 Posiciones con Mayor Relación Salario/Valor', fontsize=16)
plt.xlabel('Posición', fontsize=12)
plt.ylabel('Relación Salario/Valor', fontsize=12)

# Mostrar el gráfico
plt.show()
```

**DATA TRANSFORMATION**

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import pandas as pd
```

```python
# columnas para normalizar
columns_to_normalize = ['wage_eur', 'value_eur', 'age', 'overall']

# transformador MinMaxScaler
scaler_minmax = MinMaxScaler()

# normalización
players_16_clean[columns_to_normalize] = scaler_minmax.fit_transform(players_16_clean[columns_to_normalize])

#  verificacion
print(players_16_clean[columns_to_normalize].head())
```

```python
# transformador StandardScaler
scaler_standard = StandardScaler()


players_16_clean[columns_to_normalize] = scaler_standard.fit_transform(players_16_clean[columns_to_normalize])

print(players_16_clean[columns_to_normalize].head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualizar la distribución original de los datos
plt.figure(figsize=(12, 8))

# Subgráfico para el salario
plt.subplot(2, 2, 1)
sns.histplot(players_16_clean['wage_eur'], kde=True, color='blue')
plt.title('Distribución Salario Original')

# Subgráfico para el valor de mercado
plt.subplot(2, 2, 2)
sns.histplot(players_16_clean['value_eur'], kde=True, color='green')
plt.title('Distribución Valor de Mercado Original')

# Subgráfico para la edad
plt.subplot(2, 2, 3)
sns.histplot(players_16_clean['age'], kde=True, color='red')
plt.title('Distribución Edad Original')

# Subgráfico para la puntuación general
plt.subplot(2, 2, 4)
sns.histplot(players_16_clean['overall'], kde=True, color='orange')
plt.title('Distribución Puntuación General Original')

plt.tight_layout()
plt.show()
```

```python
# Visualizar la distribución de los datos después de la transformación
plt.figure(figsize=(12, 8))

# Subgráfico para el salario normalizado/escalado
plt.subplot(2, 2, 1)
sns.histplot(players_16_clean['wage_eur'], kde=True, color='blue')
plt.title('Distribución Salario Normalizado/Escalado')

# Subgráfico para el valor de mercado normalizado/escalado
plt.subplot(2, 2, 2)
sns.histplot(players_16_clean['value_eur'], kde=True, color='green')
plt.title('Distribución Valor de Mercado Normalizado/Escalado')

# Subgráfico para la edad normalizada/escalada
plt.subplot(2, 2, 3)
sns.histplot(players_16_clean['age'], kde=True, color='red')
plt.title('Distribución Edad Normalizada/Escalada')

# Subgráfico para la puntuación general normalizada/escalada
plt.subplot(2, 2, 4)
sns.histplot(players_16_clean['overall'], kde=True, color='orange')
plt.title('Distribución Puntuación General Normalizada/Escalada')

plt.tight_layout()
plt.show()
```
